```python
print(df.PRSN_INJRY_SEV_ID.unique())
print(df.PRSN_AIRBAG_ID.unique())
```

```python
try:
    # Merge Units_use and Primary_Person_use on "CRASH_ID"
    merged_df = pd.merge(units_df, df, on="CRASH_ID", how="inner")

    # Apply filtering conditions
    filtered_df = merged_df[
        (merged_df['PRSN_INJRY_SEV_ID'] == 'KILLED') &
        (merged_df['PRSN_AIRBAG_ID'] == 'NOT DEPLOYED') &
        (merged_df['VEH_MAKE_ID'] != 'NA')
    ]

    # Group by VEH_MAKE_ID and count occurrences
    grouped_df = filtered_df.groupby('VEH_MAKE_ID').size().reset_index(name='count')

    # Sort by count in descending order and take the top 5
    top_5_makes = grouped_df.sort_values(by='count', ascending=False).head(5)

    # Extract VEH_MAKE_ID column as a list
    result = top_5_makes['VEH_MAKE_ID'].tolist()
    print(result)

except Exception as exception:
    print(f"Error: {exception}")
```
